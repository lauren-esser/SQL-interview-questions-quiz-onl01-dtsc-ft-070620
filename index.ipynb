## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
The primary key is the column where each value is unique and identifies a single row of a table. Noted with an *. A foreign key is a column in one table that references the primary key in another column.





## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
1. One to One - Means that each record in Table A relates to one and only one record in Table B (and vice versa). Ex. two tables sharing only the EmployeeID.
2. One to Many - Means that a record in Table A can related to many records in Table B (and vice versa). Could be zero, one, or many in common. Ex. One customerID to many information on that customers orders.
3. Many to Many - Multiple reocrds in one table are associated with multiple records in another table. Ex. Each item can involve more than one Employees and each Employee can work on more than one item. 

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
1. Inner Join- returns records at an inner section of two tables
2. Left Join - returns all records from Table A and any matching from Table B. Think of appending information from Table B.
3. Right Join - returns all records from Table B and any matching from Table A. Think of appending information from Table A.
4. Full Join - Joining both tables together.

## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
GROUP BY is used to divide the rows of a table into groups that have matching values in one or more columns. We can used Aggregate functions after SELECT to summarize the groups in a more detailed manner. Ex. using SUM, COUNT(*) or DISTINCT. SUM can add up the total of the group, COUNT(*) provides a count to all the rows in regard to a field, and DISTINCT can make sure there are not duplicates.
